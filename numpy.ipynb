```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi , 200)

amplitude = 2
frequncy = 3

y = amplitude * np.sin(frequncy * x)

plt.plot(x , y)
plt.title('Modified Sin wave')
plt.xlabel("x values")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(0, 4*np.pi, 200)

# Define two waves
wave1 = np.sin(x)
wave2 = np.sin(2*x)

# Interference (combined wave)
combined = wave1 + wave2

# Plot waves
plt.plot(x, wave1, label="Wave 1", linestyle="--")
plt.plot(x, wave2, label="Wave 2", linestyle=":")
plt.plot(x, combined, label="Combined Wave", linewidth=3)

# Graph labels
plt.title("Light Wave Interference Simulation")
plt.xlabel("X values")
plt.ylabel("Amplitude")

# Legend
plt.legend()

plt.show()
```
