```python
# (Moving the Wave)
# That timing difference = phase shift.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = np.sin(x + np.pi/2)  # phase shift of π/2 (90°)
# multiply x by 2 for the higher‑frequency wave
y3 = np.sin(2 * x + np.pi/4)
y4 = np.sin(-2 * (x))
y5  = -3 *np.sin(2 * x - np.pi/2)


# plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)


plt.show()
```

-3 * → amplitude + flip

2*x → frequency

- np.pi/2 → shift right

np.linspace(...) → create smooth x values

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(0, 4*np.pi, 200)

# Define two waves
wave1 = np.sin(x)
wave2 = np.sin(2*x)

# Interference (combined wave)
combined = wave1 + wave2

# Plot waves
plt.plot(x, wave1, label="Wave 1", linestyle="--")
plt.plot(x, wave2, label="Wave 2", linestyle=":")
plt.plot(x, combined, label="Combined Wave", linewidth=3)

# Graph labels
plt.title("Light Wave Interference Simulation")
plt.xlabel("X values")
plt.ylabel("Amplitude")

# Legend
plt.legend()

plt.show()
```

```python
# Show interference of two waves in your simulation.
import numpy as np
import matplotlib.pylab as plt

x = np.linspace(0 , 4*np.pi ,500)

wave1 = np.sin(x)
wave2 = np.sin(x + np.pi/2)   #partial constructive interference.
# wave2 = np.sin(x)      # full construtor
# wave2 = np.sin(x + np.pi)      # full DIstructor

combined = wave1 + wave2

plt.plot(x ,wave1)
plt.plot(x ,wave2, linestyle="--")
plt.plot(x ,combined,linewidth=5 )

plt.title("Wave interfernce")
plt.xlabel("x")
plt.ylabel("Amlitude")

plt.show()

```

```python
# how different frequencies change the wave pattern.
import numpy as np
import matplotlib.pylab as plt

x = np.linspace(0 , 4*np.pi , 500)
wave1 = np.sin(x * 6)
wave2 = np.sin(8 * x)

result = wave1 + wave2

plt.plot(result)

plt.title("Multiple frequncy waves")
plt.xlabel("x")
plt.ylabel("Amplitude")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 1000)

# Define multiple waves
wave1 = np.sin(x)          # frequency 1, phase 0
wave2 = 0.5*np.sin(2*x + np.pi/3)   # frequency 2, phase 60°
wave3 = 0.3*np.sin(3*x + np.pi/2)   # frequency 3, phase 90°

# Resulting interference
result = wave1 + wave2 + wave3

# Plot all waves
plt.plot(x, wave1, label="Wave 1")
plt.plot(x, wave2, label="Wave 2")
plt.plot(x, wave3, label="Wave 3")
plt.plot(x, result, label="Resulting Wave", color="black", linewidth=2)

plt.title("Complex Wave Interference")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

Waves in Photonics

In photonics, light behaves like a wave. Key properties:

Amplitude → brightness of light

Frequency → color of light (red = low frequency, blue = high frequency)

Phase → determines how light waves align with each other

Constructive interference → waves add → brighter light

Destructive interference → waves cancel → dimmer or dark spot
