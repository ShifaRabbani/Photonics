```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 1000)

# Define multiple waves
wave1 = np.sin(x)          # frequency 1, phase 0
wave2 = 0.5*np.sin(2*x + np.pi/3)   # frequency 2, phase 60°
wave3 = 0.3*np.sin(3*x + np.pi/2)   # frequency 3, phase 90°

# Resulting interference
result = wave1 + wave2 + wave3

# Plot all waves
plt.plot(x, wave1, label="Wave 1")
plt.plot(x, wave2, label="Wave 2")
plt.plot(x, wave3, label="Wave 3")
plt.plot(x, result, label="Resulting Wave", color="black", linewidth=2)

plt.title("Complex Wave Interference")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

In photonics, light behaves like a wave. Key properties:

Frequency → color of light (red = low frequency, blue = high frequency)

Destructive interference → waves cancel → dimmer or dark spot

# Simple Fourier Transform Example
FFT tells you “there is something here”

fftfreq tells you “this something is 5 Hz, that something is 10 Hz”

```python
import numpy as np
import matplotlib.pyplot as plt

# Create time values from 0 to 1 second
t = np.linspace(0, 1, 500)

# Create a signal with two frequencies: 5 Hz and 10 Hz
signal = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*10*t)

# Apply Fast Fourier Transform
fft_result = np.fft.fft(signal)

# Create frequency values
freq = np.fft.fftfreq(len(t), t[1] - t[0])

# Plot original signal (Time Domain)
plt.figure()
plt.plot(t, signal)
plt.title("Original Signal (Time Domain)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()

# Plot only positive frequencies for a clearer graph
plt.figure()
plt.plot(freq[:250], np.abs(fft_result[:250]))
plt.title("Fourier Transform (Frequency Domain)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

# Frequency Detection using Fourier Transform

This project demonstrates how to detect frequencies in a signal using Python.

First, a signal is created using sine waves with different frequencies.
Then the Fast Fourier Transform (FFT) is applied to convert the signal from the time domain to the frequency domain.

The graph shows the frequencies present in the signal.

Tools used:
- Python
- NumPy
- Matplotlib

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Create time values (1 second with 500 samples)
t = np.linspace(0, 1, 500)

# Create a signal with two frequencies (5 Hz and 10 Hz)
signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)

# Apply Fast Fourier Transform (FFT)
fft_result = np.fft.fft(signal)

# Create frequency values
freq = np.fft.fftfreq(len(t), t[1] - t[0])

# Plot the original signal (time domain)
plt.figure()

plt.subplot(2,1,1)
plt.plot(t, signal)
plt.title("Signal in Time Domain")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot the frequency spectrum (frequency domain)
plt.subplot(2,1,2)
plt.plot(freq, np.abs(fft_result))
plt.title("Frequency Spectrum using FFT")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)

# Show plots
plt.tight_layout()
plt.show()
```

The frequency spectrum is simply a graph that shows which frequencies are present in a signal and how strong they are.

Time domain → how the signal changes over time

Frequency domain → what frequencies make the signal
